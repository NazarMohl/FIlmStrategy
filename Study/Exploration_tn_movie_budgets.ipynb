**Let's sort and look for duplicate records and understand what these duplicates represent**

```python
tn_movie_raw_list = tn_df['movie']
len(tn_movie_raw_list)
```

```python
tn_movie_list = []
for item in tn_movie_raw_list:
    if item[0] != '#':
        tn_movie_list.append(item)
len(tn_movie_list)
```

```python
len(sorted(set(tn_movie_list)))
```

```python
tn_movie_list[0:9]
```

```python
dup = []

tn_list_count = []
for item in tn_movie_list:
    tn_list_count.append(0)

tn_movie_set = dict(zip(sorted(tn_movie_list), tn_list_count))
tn_movie_set.keys()
                    
for key in tn_movie_set.keys():
    for movie in tn_movie_list:
        if key == movie:
            tn_movie_set[key] += 1
            if tn_movie_set[key] > 1:
#                print(key, tn_movie_set[key])
                dup.append(key)

len(dup)
```

**Randomly looking at duplicates, it is clear that these "duplicate" records are actually records of distinct movies with the same name.**

**Add column with just the release year instead of full date. Use position of comma in date to locate year.**

```python
len(tn_df)
```

```python
year=[]
for row in tn_df['release_date']:
    pos_init = row.find(',', 0, 12) + 2
    pos_end = pos_init + 4
    year.append(int(row[pos_init:pos_end]))
#     print(row[pos_init:pos_end])
# year
tn_df['year'] = year
```
